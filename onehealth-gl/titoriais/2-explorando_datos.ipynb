# Explorando datos

Para comezar, importaremos algúns módulos que en xeral nos serán de utilidade:

```python
import onehealth
import pyspark.sql.functions as F
from pyspark.sql.functions import col, expr
import pandas as pd
```

Para obter máis información sobre un conxunto de datos concreto podemos usar a función `describe`:

```python
onehealth.describe("especies/ieo_capturas")
```

Para facer exploracións cos datos, cargaremos o dataframe nunha variable para traballar máis cómodamente.

```python
capturas = onehealth.load("especies/ieo_capturas")
```

Agora que temos os datos cargados, xa podemos traballar con eles, por exemplo podemos ver cantas filas hai no conxunto de datos:

```python
capturas.count()
```

Ou ver explorar as primeiras filas do conxunto:

```python
capturas.limit(5).toPandas()
```

## Filtrando datos

Imos cargar o dataframe de `capturas` para filtrar os datos e quedarnos só cos datos de captura para o *Bacalao do Atlántico* (`cod_ieem 3582`).

```python
bacalao = capturas.filter(col('spanish_fao_name') == 'Bacalao del Atlántico')
```

Consultamos cantas filas ten o conxunto xa filtrado.

```python
bacalao.count()
```

Observamos o filtrado realizado.

```python
bacalao.limit(5).toPandas()
```
