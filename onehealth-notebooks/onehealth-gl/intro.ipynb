## 2. Módulo onehealth

Os notebooks desta plataforma están equipados co módulo `onehealth`, que proporciona acceso a unha variedade de conxuntos de datos e ferramentas para realizar análises de datos nunha contorna interactiva.

### Módulo `onehealth`

O módulo **onehealth** ofréceche as seguintes funcionalidades:

1. **list()**: Mostra unha lista de conxuntos de datos dispoñibles xunto con información detallada, como a fonte, URL, data de inclusión e descrición.
2. **load('nome_dataset')**: Carga un conxunto de datos en Spark para a súa análise. Só proporciona o nome do conxunto de datos que desexes cargar.
3. **describe('nome_dataset')**: Mostra o esquema do conxunto de datos especificado.
