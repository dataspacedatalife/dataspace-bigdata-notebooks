# Cargando datos

Para comezar, importaremos o módulo onehealth para usar todas as funcionalidades:

```python
import onehealth
import pyspark.sql.functions as F
from pyspark.sql.functions import col, expr
import pandas as pd
```

Agora podemos listar os conxuntos de datos dispoñibles coa función `list`:

```python
onehealth.list()
```

Se queremos ter máis información sobre un conxunto de datos concreto podemos usar a función `describe`:

```python
onehealth.describe("especies/ieo_capturas")
```

Finalmente podemos cargar o conxunto de datos que nos interese usando a función `load`:

```python
capturas = onehealth.load("especies/ieo_capturas")
```

Agora que temos os datos cargados, xa podemos traballar con eles, por exemplo podemos ver cantas filas hai no conxunto de datos:

```python
capturas.count()
```

Ou ver explorar as primeiras filas do conxunto:

```python
capturas.head(5)
```

```python
capturas.limit(5).toPandas()
```

# Cargando datos filtrados

> **⚠️ ADVERTENCIA**: Esta función non está actualmente dispoñible para todos os conxuntos de datos. Para calquera dúbida ou consulta, enviar unha petición a onehealth@cesga.es.

Facendo uso da función `load_partition` pódese seleccionar unha franxa de tempo a cargar.

```python
roms_filtrado = onehealth.load_partition("modelos/meteogalicia_roms", 2014, 6, 2)
```

Mostramos as primeiras filas do resultado:

```python
roms_filtrado.head(5)
```

```python
roms_filtrado.limit(5).withColumn("ocean_time", col("ocean_time").cast("string")).toPandas()
```
