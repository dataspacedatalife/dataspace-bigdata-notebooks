## 2. Módulo onehealth

Los notebooks de esta plataforma están equipados con el módulo `onehealth`, que proporciona acceso a una variedad de conjuntos de datos y herramientas para realizar análisis de datos en un entorno interactivo.

### Módulo `onehealth`

El módulo **onehealth** ofrece las siguientes funcionalidades:

1. **list()**: muestra una lista de conjuntos de datos disponibles junto con información detallada, como la fuente, URL, fecha de inclusión y descripción.
2. **load('nombre_dataset')**: carga un conjunto de datos en Spark para su análisis. Solo proporciona el nombre del conjunto de datos que desees cargar.
3. **describe('nombre_dataset')**: muestra el esquema del conjunto de datos especificado.
